```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
rmse_lnr_1 = np.sqrt(mean_squared_error(y_test1,y_pred_lnr1))
rmse_lnr_2 = np.sqrt(mean_squared_error(y_test2,y_pred_lnr2))
rmse_lnr_3 = np.sqrt(mean_squared_error(y_test3,y_pred_lnr3))
rmse_lnr_4 = np.sqrt(mean_squared_error(y_test4,y_pred_lnr4))
rmse_lnr_5 = np.sqrt(mean_squared_error(y_test5,y_pred_lnr5))
```

NSE (Nash-Sutcliffe Efficiency)

```python
nse_lnr_1 = 1 - (sum((y_test1 - y_pred_lnr1) ** 2) / sum((y_test1 - np.mean(y_test1)) ** 2))
nse_lnr_2 = 1 - (sum((y_test2 - y_pred_lnr2) ** 2) / sum((y_test2 - np.mean(y_test2)) ** 2))
nse_lnr_3 = 1 - (sum((y_test3 - y_pred_lnr3) ** 2) / sum((y_test3 - np.mean(y_test3)) ** 2))
nse_lnr_4 = 1 - (sum((y_test4 - y_pred_lnr4) ** 2) / sum((y_test4 - np.mean(y_test4)) ** 2))
nse_lnr_5 = 1 - (sum((y_test5 - y_pred_lnr5) ** 2) / sum((y_test5 - np.mean(y_test5)) ** 2))
```

```python
nse_knn_1 = 1 - (sum((y_test1 - y_pred_knn1) ** 2) / sum((y_test1 - np.mean(y_test1)) ** 2))
nse_knn_2 = 1 - (sum((y_test2 - y_pred_knn2) ** 2) / sum((y_test2 - np.mean(y_test2)) ** 2))
nse_knn_3 = 1 - (sum((y_test3 - y_pred_knn3) ** 2) / sum((y_test3 - np.mean(y_test3)) ** 2))
nse_knn_4 = 1 - (sum((y_test4 - y_pred_knn4) ** 2) / sum((y_test4 - np.mean(y_test4)) ** 2))
nse_knn_5 = 1 - (sum((y_test5 - y_pred_knn5) ** 2) / sum((y_test5 - np.mean(y_test5)) ** 2))
```

```python
nse_crt_1 = 1 - (sum((y_test1 - y_pred_reg1) ** 2) / sum((y_test1 - np.mean(y_test1)) ** 2))
nse_crt_2 = 1 - (sum((y_test2 - y_pred_reg2) ** 2) / sum((y_test2 - np.mean(y_test2)) ** 2))
nse_crt_3 = 1 - (sum((y_test3 - y_pred_reg3) ** 2) / sum((y_test3 - np.mean(y_test3)) ** 2))
nse_crt_4 = 1 - (sum((y_test4 - y_pred_reg4) ** 2) / sum((y_test4 - np.mean(y_test4)) ** 2))
nse_crt_5 = 1 - (sum((y_test5 - y_pred_reg5) ** 2) / sum((y_test5 - np.mean(y_test5)) ** 2))
```

```python
nse_svr_1 = 1 - (sum((y_test1 - y_pred_svr1) ** 2) / sum((y_test1 - np.mean(y_test1)) ** 2))
nse_svr_2 = 1 - (sum((y_test2 - y_pred_svr2) ** 2) / sum((y_test2 - np.mean(y_test2)) ** 2))
nse_svr_3 = 1 - (sum((y_test3 - y_pred_svr3) ** 2) / sum((y_test3 - np.mean(y_test3)) ** 2))
nse_svr_4 = 1 - (sum((y_test4 - y_pred_svr4) ** 2) / sum((y_test4 - np.mean(y_test4)) ** 2))
nse_svr_5 = 1 - (sum((y_test5 - y_pred_svr5) ** 2) / sum((y_test5 - np.mean(y_test5)) ** 2))
```

RSR (Root Mean Square Ratio)

```python
rsr_lnr_1 = rmse_lnr_1 / np.std(y_test1)
rsr_lnr_2 = rmse_lnr_2 / np.std(y_test2)
rsr_lnr_3 = rmse_lnr_3 / np.std(y_test3)
rsr_lnr_4 = rmse_lnr_4 / np.std(y_test4)
rsr_lnr_5 = rmse_lnr_5 / np.std(y_test5)
```

```python
rsr_knn_1 = rmse_knn_1 / np.std(y_test1)
rsr_knn_2 = rmse_knn_2 / np.std(y_test2)
rsr_knn_3 = rmse_knn_3 / np.std(y_test3)
rsr_knn_4 = rmse_knn_4 / np.std(y_test4)
rsr_knn_5 = rmse_knn_5 / np.std(y_test5)
```

```python
rsr_crt_1 = rmse_crt_1 / np.std(y_test1)
rsr_crt_2 = rmse_crt_2 / np.std(y_test2)
rsr_crt_3 = rmse_crt_3 / np.std(y_test3)
rsr_crt_4 = rmse_crt_4 / np.std(y_test4)
rsr_crt_5 = rmse_crt_5 / np.std(y_test5)
```

```python
rsr_svr_1 = rmse_svr_1 / np.std(y_test1)
rsr_svr_2 = rmse_svr_2 / np.std(y_test2)
rsr_svr_3 = rmse_svr_3 / np.std(y_test3)
rsr_svr_4 = rmse_svr_4 / np.std(y_test4)
rsr_svr_5 = rmse_svr_5 / np.std(y_test5)
```
